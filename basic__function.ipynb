```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread('/content/Birds_culture-1-copy.jpeg')
# image_array = np.asarray(bytearray(img), dtype=np.uint8)
# image = cv.imdecode(image_array, cv.IMREAD_COLOR)
plt.imshow(img)
plt.show()

# #converting to gray scale
# gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# plt.imshow("Gray scale image",gray)
# plt.show()
```

```python
#converting to gray scale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.show()
```

```python
#converting Blur
blur = cv.GaussianBlur(img,(9,9),cv.BORDER_DEFAULT)
plt.imshow(blur)
```

```python
#Edge Blur
canny = cv.Canny(img,130,175)
plt.imshow(canny)
```

```python
#Dilating the image
dilated = cv.dilate(canny,(8,8),iterations = 3)
plt.imshow(dilated)
```

```python
#Eroding

croded = cv.erode(dilated,(7,7),iterations = 3)
plt.imshow(croded)
```

```python
#REsize

resized = cv.resize(img,(500,500),interpolation=cv.INTER_CUBIC)
plt.imshow(resized)
```

```python
#cropping
cropped = img[50:200,200:400]
plt.imshow(cropped)
# #cropping
# cropped = img[100:200,400:500]
# plt.imshow(cropped)
```
