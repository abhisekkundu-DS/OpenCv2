```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("/content/Birds_culture-1-copy.jpeg")
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title("Gray scale")
```

```python
#simple thresholding
threshold, thresh = cv.threshold(gray,150,255,cv.THRESH_BINARY)
plt.imshow(thresh)
plt.title("simple Threshold")
```

```python
threshold, thresh_inv = cv.threshold(gray,150,225,cv.THRESH_BINARY_INV)
plt.imshow(thresh_inv)
plt.title("simple Threshold inverse")
```

```python
#Adptive thresholding
adaptive_thresh = cv.adaptiveThreshold(gray,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY_INV,11,9)
plt.imshow(adaptive_thresh)
plt.title("Adptive Threshold")
```
