```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank = np.zeros((500,500,3), dtype='uint8')
plt.imshow( blank)
plt.title("Blank")
```

```python
#print the image certain color
blank[200:300,300:400]=0,0,250
plt.imshow(blank)
```

```python
blank[200:300,300:400]=0,250,0
plt.subplot(131),plt.imshow(blank)

blank[200:300,300:400]=250,0,0
plt.subplot(132),plt.imshow(blank)

blank[200:300,300:400]=0,0,250
plt.subplot(133),plt.imshow(blank)
plt.show()
```

```python
#print a rectangle
cv.rectangle(blank,(0,0),(250,250),(0,255,0),thickness=2)
plt.subplot(131),plt.imshow(blank)

cv.rectangle(blank,(500,500),(250,250),(0,255,0),thickness=2)
plt.subplot(132),plt.imshow(blank)
```

```python
#draw a circle
cv.circle(blank,(blank.shape[1]//2,blank.shape[0]//2),40,(0,0,255),thickness=-1)
plt.subplot(131),plt.imshow(blank)
```

```python
# 5. Write text
cv.putText(blank, 'Hello, my name is ABHISEK!!!', (0,225), cv.FONT_HERSHEY_TRIPLEX, 1.0, (0,255,0), 2)
plt.imshow( blank)
```
